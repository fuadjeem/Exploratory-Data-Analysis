## Handling missing values in pandas

```python
""" we will explore how we can use various pandas techniques to handle the
missing data in our datasets. We will learn how to find out how much data is missing, and
from which columns. We'll see how to drop rows or columns where all or a lot of records
are missing data. We'll also learn how, instead of dropping data, we can also fill in the
missing records with zeros or the mean of the remaining values """
```

```python
data = pd.read_csv('data-titanic.csv')
data.head()
```

```python
data.shape #We can see that the total number of records is 891, and that the total number of columns is 12.
```

```python
data.count()
""" The difference between the total records and the count per column represents the number
of records missing from that column. Out of the 12 columns, we have 3 columns where
values are missing. For example, Age has only 714 values out of a total of 891 rows; Cabin
has values for only 204 records; and Embarked has values for 889 records. There are
different ways we could handle these missing values. One of the ways is to drop any row
where a value is missing, even from a single column, as follows: """
```

```python
data_missing_dropped = data.dropna()
data_missing_dropped.shape
""" When we run this drop rows method, we assign the results back into a new DataFrame.
This leaves us with just 183 records, out of a total of 891. However, this may lead to losing
a lot of the data, and may not be acceptable. """
```

```python
""" Another method is to drop only those rows where all values are missing. Here is an
example: """
data_all_missing_dropped = data.dropna(how="all")
data_all_missing_dropped.shape
```

```python
""" Instead of dropping rows, another method is to fill in the missing values with some data.
We can fill in the missing values with 0, for example, as in the following code: """
data_filled_zeros = data.fillna(0)
data_filled_zeros.count()
```

```python
""" Also, instead of filling in missing values with 0, we could fill them with the mean of the
remaining existing values. To do this, we call the fillna method on the column where we
are filling the values in, and we pass the mean of the column as the parameter: """
data_filled_in_mean = data.copy()
data_filled_in_mean.Age.fillna(data.Age.mean(), inplace= True)
data_filled_in_mean.count()
```

## Merging and concatenating multiple DataFrames into one

```python
import pandas as pd

# Create the first dataframe
data1 = {'ID': [1, 2, 3, 4, 5],
         'Name': ['John', 'Emily', 'Ryan', 'Emma', 'Sophia'],
         'Age': [25, 30, 35, 40, 28]}
df1 = pd.DataFrame(data1)

# Create the second dataframe
data2 = {'ID': [1, 2, 5, 6, 7],
         'City': ['New York', 'London', 'Paris', 'Sydney', 'Berlin'],
         'Salary': [50000, 60000, 70000, 55000, 65000]}
df2 = pd.DataFrame(data2)



```

```python
df1
```

```python
df2
```

```python
# Merge the dataframes based on the 'ID' column
merged_df = pd.merge(df1, df2, on='ID', how='inner')
print(merged_df)
```

### While merge() is used for combining dataframes based on common columns or indices, concat() is used for concatenating dataframes along a particular axis (either rows or columns).

```python


# Create the first dataframe
data1 = {'ID': [1, 2, 3],
         'Name': ['John', 'Emily', 'Ryan']}
df1 = pd.DataFrame(data1)

# Create the second dataframe
data2 = {'ID': [4, 5, 6],
         'Name': ['Emma', 'Sophia', 'David']}
df2 = pd.DataFrame(data2)
df1

```

```python
df2
```

```python
# Concatenate the dataframes vertically (along rows)
concatenated_df = pd.concat([df1, df2], axis=0)
print(concatenated_df)
```

<!-- If you want to concatenate columns from multiple dataframes, you can use the concat() function in pandas with axis=1 to concatenate the columns horizontally. Here's an example: -->

```python
import pandas as pd

# Create the first dataframe
data1 = {'ID': [1, 2, 3],
         'Name': ['John', 'Emily', 'Ryan']}
df1 = pd.DataFrame(data1)

# Create the second dataframe
data2 = {'Age': [25, 30, 35],
         'City': ['New York', 'London', 'Paris']}
df2 = pd.DataFrame(data2)

df1
```

```python
df2
```

```python
# Concatenate the columns horizontally
concatenated_df = pd.concat([df1, df2], axis=1)
print(concatenated_df)
```

**Note that the concatenation is performed based on the index of the dataframes. If the index values are not aligned, NaN values will be inserted in the resulting dataframe. To ensure alignment, you can reset the index before concatenating or use the ignore_index=True parameter in the concat() function.**

**If there are rows in one dataframe that do not have a matching ID in the other dataframe, an inner merge will exclude those rows from the merged result. Here's an example to illustrate this scenario:**

```python
import pandas as pd

# Create the first dataframe
data1 = {'ID': [1, 2, 3, 4],
         'Name': ['John', 'Emily', 'Ryan', 'Emma'],
         'Age': [25, 30, 35, 40]}
df1 = pd.DataFrame(data1)

# Create the second dataframe
data2 = {'ID': [2, 3, 5, 6],
         'City': ['London', 'Paris', 'Sydney', 'Berlin'],
         'Salary': [60000, 70000, 55000, 65000]}
df2 = pd.DataFrame(data2)

df1
```

```python
df2
```

```python
# Perform an inner merge
merged_df = pd.merge(df1, df2, on='ID', how='inner')
print(merged_df)
```

**If you perform an outer merge using the merge() function with how='outer', it will include all the rows from both dataframes, regardless of whether there is a match on the merge key. Here's an example:**

```python
import pandas as pd

# Create the first dataframe
data1 = {'ID': [1, 2, 3, 4],
         'Name': ['John', 'Emily', 'Ryan', 'Emma'],
         'Age': [25, 30, 35, 40]}
df1 = pd.DataFrame(data1)

# Create the second dataframe
data2 = {'ID': [2, 3, 5, 6],
         'City': ['London', 'Paris', 'Sydney', 'Berlin'],
         'Salary': [60000, 70000, 55000, 65000]}
df2 = pd.DataFrame(data2)
df1
```

```python
df2
```

```python

# Perform an outer merge
merged_df = pd.merge(df1, df2, on='ID', how='outer')
print(merged_df)
# note that Where there is a match on the merge key ('ID'), the columns are combined. 
# For the unmatched rows, NaN (missing) values are inserted in the corresponding columns.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display
import seaborn as sns
```

```python

# Filter function
def filter_countries(starting_letter):
    filtered_df = df[df['country'].str.startswith(starting_letter)]
    
    plt.figure(figsize=(12, 6))
    sns.swarmplot(x='country', y='wine_servings', data=filtered_df)
    plt.xticks(rotation=45)
    plt.show()

# Create the interactive dropdown widget
letters = [chr(i) for i in range(65, 91)]  # A-Z
dropdown = widgets.Dropdown(options=letters, description='Starting Letter:')
output = widgets.interactive_output(filter_countries, {'starting_letter': dropdown})

# Display the dropdown and output
display(dropdown, output)
```
